```python
import msaf
import librosa
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

folder_path = r'C:\Users\david\fall2024\capstone\Wav'
wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
wav_files_sorted = sorted(wav_files, key=lambda x: int(x.split('_')[0]))
song_paths = [os.path.join(folder_path, f) for f in wav_files_sorted]

for path in song_paths:
    print(path)
```

```python
# excluded olda boundary algorithm, doesn't work
boundary_algos = ['cnmf', 'foote', 'scluster', 'sf', 'vmo', 'cbm']
label_algos = msaf.get_all_label_algorithms()
```

```python
for b in boundary_algos:
    for l in label_algos:
        boundaries, labels = msaf.process(song_paths[0], boundaries_id=b, labels_id=l)
        print(b, l)
        print(boundaries)
        print(len(boundaries))
        print(labels)
        print(len(labels))
```

```python
folder_destination = 'C:/Users/david/fall2024/capstone/msaf_dataset'

for s in song_paths:
    song_name = os.path.basename(s).split('.')[0]
    song_folder = os.path.join(folder_destination, song_name)
    os.makedirs(song_folder, exist_ok=True)

    for b in boundary_algos:
        algo_folder = os.path.join(song_folder, b)
        os.makedirs(algo_folder, exist_ok=True)

        for l in label_algos:
            boundaries, labels = msaf.process(s, boundaries_id=b, labels_id=l)
            boundaries = boundaries[1:]

            df = pd.DataFrame({
                'boundaries': boundaries,
                'labels': labels
                })
            
            output_file_name = f"{song_name}_{b}_{l}.csv"
            output_path = os.path.join(algo_folder, output_file_name)
            df.to_csv(output_path, index=False)
```
